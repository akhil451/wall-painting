```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('image.jpeg')
```

```python
plt.imshow(image[:,:,::-1])
```

```python
img=image.copy()
print(img.shape)
canny=cv2.Canny(img,60,25) # thresholds for initial strong edge detection and the threshold for linking weak edges, respectively
laplacian = cv2.Laplacian(img,cv2.CV_64F)   
# im_floodfill= img.copy()
h, w = img.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
# mask =  cv2.bitwise_not(mask)
im_floodfill=cv2.floodFill(canny,mask, (0,0), (100,100,255),(10,)*3, (10,)*3,4)[1]
im_floodfill=cv2.bitwise_not(im_floodfill)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
dilation = cv2.dilate(im_floodfill,kernel,iterations = 1)
# print(im_floodfill.shape)
```

```python
im_floodfill.shape
```

```python
plt.imshow(im_floodfill,cmap='gray')
print(dilation.shape)
```

```python
erosion_kernel=np.uint8(np.zeros((30,30)))
erosion_kernel[:,::5]=1
erosion_kernel[::5]=1
# print(erosion_kernel)
# dilation_gray = cv2.cvtColor(dilation,cv2.COLOR_BGR2GRAY)
# dilation= np.uint8(dilation)
print(erosion.shape)
erosion=cv2.erode(dilation,erosion_kernel,iterations=1)
plt.imshow(erosion,cmap='gray')
```
